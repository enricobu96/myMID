```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('xy_ensemble.csv')

plt.rcParams['figure.figsize'] = [15, 15]
plt.title('Ensemble Losses')
plt.xlabel('Iteration')
plt.ylabel('Log Loss')
plt.plot(df['iterations'], df['log_loss'], label='log_loss')
plt.plot(df['iterations'], df['log_fake_loss'], label='MSE train-samples')
plt.plot(df['iterations'], df['kde_ll'], label='KDE NLL')
```
